```python
# Escalado de variables
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cc_df_scaled = scaler.fit_transform(cc_df)
```

<h1 style="color: green;"> Elección de cantidad de 
    Clusters </h1>

```python
from sklearn.cluster import KMeans

score_1 = []
for i in range(1,21):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(cc_df_scaled)
    score_1.append(kmeans.inertia_) #WCSS
```

```python
# Visualización del codo
plt.plot(range(1,21), score_1, 'gp-')
plt.title('Encontrando el npumero óptimo de clusters')
plt.xlabel('Cantidad de clusters')
plt.ylabel('WCSS(k)')
plt.show()

# Al parecer no se ha formado un "codo" muy pronunciado aun asi se observa que desde 6 a 10 clústers es lo óptimo
# Usaremos 8 aunque mas tarde con GrindSearchCV probaremos el rango dicho anteriormente.
```

```python
# aplicamos kmeans con la cantidad óptima
kmeans = KMeans(8)
kmeans.fit(cc_df_scaled)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
cluster_centers = pd.DataFrame(data= kmeans.cluster_centers_, columns=[cc_df.columns])
cluster_centers

# Podemos ver donde las coordenadas de cada centroide, cada columna es una característica y cada característica 
# es una dimensión y un punto de coordenada
```

<h1 style="color: green"> Reducción de la Dimensionalidad con Autoencoders</h1>

```python
from tensorflow.keras.layers import Input, Add, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D, AveragePooling2D, MaxPooling2D, Dropout
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.initializers import glorot_uniform
from tensorflow.keras.optimizers import SGD
```

```python
# cantidad de dimension a comprimir
encoding_dimension = 7

# tamaño de entrada de la RNA
input_df = Input(shape=(17,))

# Codificación

# Primer capa
x = Dense(encoding_dimension, activation ='relu')(input_df) 

# Segunda capa
x = Dense(500, activation='relu', kernel_initializer= glorot_uniform)(x)

# Tercer capa
x = Dense(500, activation='relu', kernel_initializer= glorot_uniform)(x)

# Cuarta capa
x = Dense(2000, activation='relu', kernel_initializer= glorot_uniform)(x)

# Quinta capa
encoded = Dense(10,activation='relu', kernel_initializer= glorot_uniform)(x)


# Decodificación

# Primer capa
x = Dense(2000, activation='relu', kernel_initializer= glorot_uniform)(encoded)

# Segunda capa
x = Dense(500, activation='relu', kernel_initializer= glorot_uniform)(x)

# Tercer capa
decoded = Dense(17, kernel_initializer= glorot_uniform)(x)


autoencoder = Model(input_df, decoded)

encoder = Model(input_df, encoded) 


autoencoder.compile(optimizer='adam', loss = 'mean_squared_error')
```

```python
autoencoder.summary()
```

```python
autoencoder.fit(cc_df_scaled, cc_df_scaled, batch_size=128, epochs=25, verbose=1)
```
